```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
file_path = r"O:\AI_ML\Github\InsightGen-AI\data\netflix_titles.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
# Define model

from langchain_ollama.llms import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

template = """You are an expert data analyst and Python developer.

Your task is to analyze a pandas DataFrame called `df` and answer the user's question.

{columns} These are our columns in df

Rules:
1. You MUST only use the dataframe named `df`.
2. Do NOT read or write files.
3. Do NOT import new libraries unless explicitly required.
4. Use only pandas, numpy, matplotlib, or seaborn.
5. If a visualization is required, generate Python plotting code.
6. Always return valid and executable Python code.
7. Do NOT explain the code unless explicitly asked.
8. If the query cannot be answered from the data, say: 
   "The requested analysis cannot be performed with the available data."

Output format:
- First output: Python code only
- No markdown
- No explanations"""

columns = df.columns.tolist()

prompt = ChatPromptTemplate.from_template(template)

llm = OllamaLLM(model="llama3.1:8b", streaming=True)

chain = prompt | llm
```

```python
query = "Explain pokemon go AR mode"
for chunk in llm.stream(query):
    print(chunk, end="", flush=True)
```
